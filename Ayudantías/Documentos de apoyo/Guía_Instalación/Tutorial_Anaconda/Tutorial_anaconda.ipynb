Ahora bien, muchas veces nos interesa trabajar con una versin en particular de python porque un proyecto est hecho con una determinada versin de una librera y esa versin es compatible con una versin previa de Python. 

Para solucionar esto, Anaconda nos permite crear instancias de python con sus propias versiones de librerias, las cuales se llaman **ambientes (Enviroments)**

**Les recomiendo que se creen su propio ambiente en Anaconda, lo cual en realidad es muy simple.**

Para crear un ambiente, una vez instalado Anaconda, debemos abrir la terminal (Linux/Mac) o la Anaconda prompt (Windows) y hacer lo siguiente:

* conda create --name **nombre_ambiente** python=3.x -y (Puede ser la versin de python que estimen conveniente. Recomiendo la ms actual)

Esto nos va a cargar en la terminal nuestro ambiente. Luego debemos activarlo (Siempre que vayamos a usarlo debemos activarlo)
* conda activate **nombre_ambiente** 

Una vez creado el ambiente, instalamos las libreras que usaremos durante el semestre:
* pip install **libreria** 



Ahora bien, para fortuna de nuestro tiempo, tenemos un archivo requeriments.txt. Este archivo ya contiene todas las libreras que utilizaremos y solo debemos ejecutarlo en nuestro ambiente:
* pip install -r requirements.txt
